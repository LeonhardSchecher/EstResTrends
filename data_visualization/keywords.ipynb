```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json('keywords.json')

keywords = df['keyword'].explode()
freq = keywords.value_counts()

freq_filtered = freq[freq >= 75]

ax = freq_filtered.plot(
    kind='bar',
    fontsize=10,
    rot=85,
    title='Keyword frequencies (75 times or more)'
)

ax.set_xlabel('Keyword')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
data = pd.read_json('data.json')

filter = data[data['Keywords'].notna()]
freq = filter['Year'].value_counts()


ax = freq.plot(
    kind='bar',
    fontsize=10,
    rot=85,
    title='Year'
)

ax.set_xlabel('Year')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```
