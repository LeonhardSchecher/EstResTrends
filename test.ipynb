```python
import os
def log_token_usage(save_path:str, current_usage:int):
    if not os.path.exists(save_path):
        with open(save_path, 'x') as f:
            f.write('0')
    with open(save_path, 'r') as f:
        old_usage = int(f.read())
    with open(save_path, 'w') as f:
        f.write(str(old_usage+current_usage))
```
